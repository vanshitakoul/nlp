Morphology is the study of the way words are built up from smaller meaning bearing units. Study
and understand the concepts of morphology by the use of add delete table

```python
# ==========================================
# MORPHOLOGY USING ADD-DELETE TABLE IN NLP
# ==========================================

import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download package
nltk.download('wordnet')

# Initialize tools
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Input
word = input("Enter a word: ")

print("\n--- MORPHOLOGY USING ADD-DELETE TABLE ---")

# Prefix
print("Add Prefix 'un' :", "un" + word)

# Add 'ing'
if word.endswith("e"):
    print("Delete 'e' + ing :", word[:-1] + "ing")
else:
    print("Add 'ing'        :", word + "ing")

# Add 'ness'
if word.endswith("y"):
    print("y → i + ness     :", word[:-1] + "iness")
else:
    print("Add 'ness'       :", word + "ness")

# Add 'er'
if word.endswith("y"):
    print("y → i + er       :", word[:-1] + "ier")
else:
    print("Add 'er'         :", word + "er")

# Stemming
print("\nStemmed Word     :", stemmer.stem(word))

# Lemmatization
print("Lemmatized Word  :", lemmatizer.lemmatize(word))
```
