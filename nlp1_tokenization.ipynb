```python
# Import libraries
import nltk
from nltk.tokenize import word_tokenize, wordpunct_tokenize
from nltk.tokenize import TreebankWordTokenizer
from nltk.tokenize import TweetTokenizer
from nltk.tokenize import MWETokenizer

from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer

# Download required datasets
nltk.download('punkt')
nltk.download('wordnet')

# Sample text
text = "Hello everyone! I am learning NLP using NLTK library. #AI is amazing :)"

print("Original Text:\n", text)

# ---------------------------------------------------
# 1. Whitespace Tokenization
# ---------------------------------------------------
print("\n1. Whitespace Tokenization:")
whitespace_tokens = text.split()
print(whitespace_tokens)

# ---------------------------------------------------
# 2. Punctuation-based Tokenization
# ---------------------------------------------------
print("\n2. Punctuation-based Tokenization:")
punct_tokens = wordpunct_tokenize(text)
print(punct_tokens)

# ---------------------------------------------------
# 3. Treebank Tokenization
# ---------------------------------------------------
print("\n3. Treebank Tokenization:")
treebank = TreebankWordTokenizer()
treebank_tokens = treebank.tokenize(text)
print(treebank_tokens)

# ---------------------------------------------------
# 4. Tweet Tokenization
# ---------------------------------------------------
print("\n4. Tweet Tokenization:")
tweet = TweetTokenizer()
tweet_tokens = tweet.tokenize(text)
print(tweet_tokens)

# ---------------------------------------------------
# 5. MWE (Multi Word Expression) Tokenization
# ---------------------------------------------------
print("\n5. MWE Tokenization:")

mwe_tokenizer = MWETokenizer([('New', 'York'), ('machine', 'learning')], separator='_')

sentence = "I live in New York and study machine learning"

mwe_tokens = mwe_tokenizer.tokenize(sentence.split())
print(mwe_tokens)

# ---------------------------------------------------
# Stemming
# ---------------------------------------------------
words = ["playing", "played", "plays", "happiness"]

# Porter Stemmer
print("\nPorter Stemmer:")
porter = PorterStemmer()

for word in words:
    print(word, "->", porter.stem(word))

# Snowball Stemmer
print("\nSnowball Stemmer:")
snowball = SnowballStemmer("english")

for word in words:
    print(word, "->", snowball.stem(word))

# ---------------------------------------------------
# Lemmatization
# ---------------------------------------------------
print("\nLemmatization:")

lemmatizer = WordNetLemmatizer()

for word in words:
    print(word, "->", lemmatizer.lemmatize(word))
```
