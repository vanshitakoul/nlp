```python
# Text Cleaning, Lemmatization, Stopword Removal, Label Encoding, TF-IDF

# Install required libraries (Run once)
!pip install pandas scikit-learn nltk

import pandas as pd
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer

# Download NLTK resources (Run once)
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# ---------------------------------------------------
# Sample Dataset
# ---------------------------------------------------

data = {
    "Text": [
        "I love Machine Learning and AI!",
        "NLP is very interesting.",
        "Python programming is easy.",
        "I enjoy learning new technologies."
    ],

    "Category": [
        "Positive",
        "Positive",
        "Positive",
        "Positive"
    ]
}

df = pd.DataFrame(data)

print("Original Data:\n")
print(df)

# ---------------------------------------------------
# Step 1: Text Cleaning
# ---------------------------------------------------

def clean_text(text):

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    return text

df['Clean_Text'] = df['Text'].apply(clean_text)

print("\nAfter Cleaning:\n")
print(df['Clean_Text'])

# ---------------------------------------------------
# Step 2: Remove Stopwords
# ---------------------------------------------------

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):

    words = text.split()

    filtered_words = [word for word in words if word not in stop_words]

    return " ".join(filtered_words)

df['No_Stopwords'] = df['Clean_Text'].apply(remove_stopwords)

print("\nAfter Stopword Removal:\n")
print(df['No_Stopwords'])

# ---------------------------------------------------
# Step 3: Lemmatization
# ---------------------------------------------------

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):

    words = text.split()

    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(lemmatized_words)

df['Lemmatized_Text'] = df['No_Stopwords'].apply(lemmatize_text)

print("\nAfter Lemmatization:\n")
print(df['Lemmatized_Text'])

# ---------------------------------------------------
# Step 4: Label Encoding
# ---------------------------------------------------

label_encoder = LabelEncoder()

df['Encoded_Label'] = label_encoder.fit_transform(df['Category'])

print("\nEncoded Labels:\n")
print(df['Encoded_Label'])

# ---------------------------------------------------
# Step 5: TF-IDF Representation
# ---------------------------------------------------

tfidf = TfidfVectorizer()

tfidf_matrix = tfidf.fit_transform(df['Lemmatized_Text'])

# Convert TF-IDF matrix to DataFrame
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)

print("\nTF-IDF Representation:\n")
print(tfidf_df)

# ---------------------------------------------------
# Step 6: Save Outputs
# ---------------------------------------------------

# Save processed dataset
df.to_csv("processed_text_data.csv", index=False)

# Save TF-IDF values
tfidf_df.to_csv("tfidf_output.csv", index=False)

print("\nOutputs saved successfully!")
print("1. processed_text_data.csv")
print("2. tfidf_output.csv")
```
